# Google Play Store Data Analysis

## Overview
This project analyzes Google Play Store app data to understand trends in app categories, ratings, and user engagement.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("google_playstore.csv")

# Display first few rows
df.head()
```

```python
# Checking null values
df.isnull().sum()

# Dropping rows with missing values
df = df.dropna()

# Converting Reviews column to numeric
df["Reviews"] = df["Reviews"].astype("float")

# Converting Rating column to numeric
df["Rating"] = df["Rating"].astype("float")
```

```python
df
```

```python
# Top categories
df["Category"].value_counts().head(10)

# Average rating by category
df.groupby("Category")["Rating"].mean().sort_values(ascending=False).head(10)

# Most reviewed apps
df.sort_values(by="Reviews", ascending=False).head(10)
```

```python
# Top categories visualization
df['Category'].value_counts().head(10).plot(kind='bar')
plt.title("Top App Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

## Key Insights

- Apps with higher number of reviews tend to have better visibility and popularity
- Certain categories like Games and Tools dominate the Play Store
- Ratings are generally consistent across most categories
- Free apps are more common than paid apps
